### ***Modélisation d'une portion du signal par un polynôme***

```python
def modelepoly(x,y,xMin,xMax,degre):
    """
    x     : angles d'abscisse du signal complet
    y     : amplitude du signal complet pour les valeurs des angles inclus dans x
    xMin  : abscisse de la borne min de la portion de signal que l'on prélève 
    xMax  : abscisse de la borne max de la portion de signal que l'on prélève 
    degre : degre du polynôme que l''on souhaite utiliser pour modéliser cette portion de signal
    """ 

    # Prélèvement de la portion du signal
    if xMin < min(x): 
        iMin = 0
    elif xMin > max(x):
        iMin = len(x)-1
    else:
        iMin = 0
        while (x[iMin] < xMin):
            iMin += 1
            
    if xMax < min(x):
        iMax = 0
    elif xMax > max(x):
        iMax = len(x)-1
    else:
        iMax = 0
        while (x[iMax] < xMax):
            iMax += 1           
    
    if iMin < iMax:
        x = x[iMin:iMax]
        y = y[iMin:iMax]
    else:
        print('Les bornes sont erronées, on reste sur le signal complet')
        
    # Régression polynomiale
    
    M, V = np.stack([x**k for k in range(degre+1)], axis=1), y
    yhat = M @ np.linalg.lstsq(M, V,rcond=None)[0]
    # visualisation graphique    
    plt.figure(figsize=(8,8))
    plt.scatter(x, y, label="données enregistrées")
    plt.plot(x, yhat, c='r', label="modèle des données")
    plt.xlabel('angle (en degrés)', fontsize=18)
    plt.ylabel('signal (en Volt)', fontsize=18)
    plt.legend(fontsize=18)
    plt.grid()
    plt.show()        
```

```python
modelepoly(x,y,-90,90,2)
```

```python
modelepoly(x,y,-180,180,2)
```

```python
modelepoly(x,y,-180,180,3)
```

```python
modelepoly(x,y,-180,180,4)
```

```python
modelepoly(x,y,-250,250,4)
```

Le comportement du polynôme pour $x=\pm215^\circ$ n'est pas réaliste  
**-> on peut vérifier que c'est un problème de dépassement de la précision de calcul**  
En effet, en parcourant la dernière colonne de la matrice $M$, on peut lire que $-216^4$ est négatif      
Le nombre est tellement grand, qu'il déborde de ce que peut représenter le format par défaut du nombre  
Ce débordement en fait un nombre négatif

**Une première possibilité est d'augmenter la précision de calcul**

```python
x = x.astype('float64')
modelepoly(x,y,-250,250,4)
```

```python
modelepoly(x,y,-250,250,6)
```

```python
modelepoly(x,y,-360,360,7)
```

```python
modelepoly(x,y,-360,360,8)
```

A force d'augmenter l'ordre du polynôme pour pouvoir décrire des plages de plus en plus longues,  
le problème de précision se présentera de nouveau, sans qu'on ne puisse plus augumenter cette précision

**-> Une bien meilleure approche est de normaliser toutes les données entre 0 et 1**  
De cette façon, on est alors beaucoup moins sensible aux problèmes de précision de calcul

```python
x = polySpline[polySpline.files[0]]
x = x.astype('float64')
x=x/360
modelepoly(x,y,-1,1,8)
```

```python
x = polySpline[polySpline.files[0]]
x = x.astype('float64')
x=x/720
modelepoly(x,y,-1,1,8)
```

```python
modelepoly(x,y,-1,1,12)
```

```python
modelepoly(x,y,-1,1,16)
```

Augmenter ainsi sans cesse l'ordre des polynômes n'est cependant pas désirable  
**-> une solution beaucoup plus satisfaisante est de se tourner vers des B-Splines**

### ***Modélisation d'une portion du signal par des B-Splines***

#### **Saisie des polynômes de base**

```python
def b0(t):
     return 0.5*(1-2*t+t**2)
def b1(t):
    return 0.5*(1+2*t-2*t**2)
def b2(t):
    return 0.5*t**2
```

```python
def a0(t):
     return t**2-2*t+1
def a1(t):
    return 0.5*(-3*t**2+4*t)
def a2(t):
    return 0.5*t**2
```

```python
def c0(t):
     return 0.5*(t**2-2*t+1)
def c1(t):
    return 0.5*(-3*t**2+2*t+1)
def c2(t):
    return t**2
```

Pour simplifier l'étude, on revient avec des abscisses non normalisées  
c'est-à-dire en degrés, entre -720 et 720

On définit **40 segments de 36°**, cela donnera lieu à 42 points de contrôle $P$

```python
x = polySpline[polySpline.files[0]]
```

```python
Nbseg = 40
Nbtaille = int((x[-1]-x[0])/Nbseg)
```

```python
V = y
M = np.zeros((len(x),Nbseg+2))
for i in range(len(x)):

    # on identifie le segment sur lequel se trouve x[i]                                                                                               
    seg = int((x[i] - x[0])/Nbtaille)
    # on calcule l'abscisse de x[i] sur ce segment
    t = (x[i] - x[seg*Nbtaille])  / Nbtaille
    
    
    # on remplit la matrice M
    if seg == 0:
        M[i,seg:seg+3] = np.array([a0(t), a1(t), a2(t)])
    elif seg == (Nbseg -1):
        M[i,seg:seg+3] = np.array([c0(t), c1(t), c2(t)])
    elif seg < (Nbseg -1):
        M[i,seg:seg+3] = np.array([b0(t), b1(t), b2(t)])
    else:
        M[i,seg-1:seg+2] = np.array([c0(1), c1(1), c2(1)])
        
P, _, _ ,_  = list(np.linalg.lstsq(M, V,rcond=None))
haty = M @ P
```

#### Visualisation graphique

```python
plt.figure(figsize=(8,8))
plt.scatter(x, y, alpha = 0.5)

for i in range(Nbseg):
    if i%2 == 0:
        C = 'red'
    else:
        C = 'cyan'
    plt.plot(x[i*Nbtaille:(i+1)*Nbtaille+1], 
            haty[i*Nbtaille:(i+1)*Nbtaille+1], c=C)

X = np.zeros(Nbseg+2) 
X[0] = x[0]
X[1] = x[0] + 0.5*Nbtaille
X[-1] = x[-1]
for i in range(2,Nbseg+1):
    X[i] = X[i-1] + Nbtaille 
plt.scatter(X, P, marker = 'o', c='black')    
    
plt.xlabel('angle (en degrés)', fontsize=18)
plt.ylabel('signal (en Volt)', fontsize=18)
plt.grid()
plt.show() 
```
